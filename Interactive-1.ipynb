```python
import yt
import matplotlib.pyplot as plt
import numpy as np
import trident
import outflow_rate
import fitting

def rate_ew(time_step, start, end, lines, plot = False):
    #Determine outflow rate
    ds = outflow_rate.load_data(timestep = time_step)
    rate = outflow_rate.outflow_rate(ds, 10, -10)


    '''Add ion fields, creating the light ray, and finding the equivalent width'''
    #Adding ion fields
    trident.add_ion_fields(ds, ions=['Mg II'])
    trident.add_ion_fields(ds, ions=['O VI'])
    trident.add_ion_fields(ds, ions=['C IV'])

    #Creating the light ray
    def to_cl(kpc):
        return kpc*3.08567758128E+21/(0.150000000000000E+03*0.308567758128200E+22) #have to change accordingly with the specific file
    # ray_start = [to_cl(i)+0.5 for i in start]
    # ray_end = [to_cl(i)+0.5 for i in end]
    ray_start = start
    ray_end = end
    ray = trident.make_simple_ray(ds,
                                start_position=ray_start,
                                end_position=ray_end,
                                data_filename="ray.h5",
                                lines = ['H', 'C', 'N', 'O', 'Mg', 'Si'])
    
    if plot:
        p = yt.ProjectionPlot(ds, 'y', 'density')
        p.annotate_ray(ray, arrow=True)
        p.show()
    
    #Generating spectrum
    sg = trident.SpectrumGenerator(lambda_min = 800, lambda_max = 1600, dlambda=0.02)
    sg.make_spectrum(ray, lines = 'all')
    wavelength = sg.lambda_field.value
    flux = sg.flux_field
    # print('FFFFFFFFFFFFF')
    # plt.plot(wavelength, flux)
    # plt.show()
    # print('FFFFFFFFFFFFF')
    plt.plot(wavelength, flux)
    plt.show()
    #specifying background then fit
    background = np.ones(len(flux)) #need to change later
    fig, axO, ew_fit, ew_raw = fitting.fitting(wavelength, flux, lines, bg=background)

    return (rate, ew_fit, ew_raw)

rates = []
ews_fits_1 = []
ews_fits_2 = []
ews_raws_1 = []
ews_raws_2 = []
for i in range(191, 203):
    rate, ew_fit, ew_raw = rate_ew(203, [1, 0.5, 1], [0, 0.5, 0.1], ['O VI 1032', 'O VI 1038'], plot = True)
    rates.append(rates)
    ews_fits_1.append(ew_fit['O VI 1032'])
    ews_fits_2.append(ew_fit['O VI 1038'])
    ews_raws_1.append(ew_raw['O VI 1032'])
    ews_raws_2.append(ew_raw['O VI 1038'])
```
